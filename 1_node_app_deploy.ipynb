## 1. Deploy the interactive "TODO" web app 

I now run the `azd up` command to provision and deploy the template resources to Azure. When prompted, I select the subscription and location to deploy to.

### Azure developer CLI `azd up` command breakdown

The `azd up` command runs the three separate commands in sequence:
1. `azd package` for packing web app 
2. `azd provision` for provisioning infrastructure
3. `azd deploy` for deploying web app hosted on provisioned infrastructure

The execution of `azd up` therefore will take some time. As I am running my project on a mobile network, so I expect it to be even longer.
