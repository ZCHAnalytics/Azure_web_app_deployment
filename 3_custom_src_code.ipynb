## 3. Customising web application code 

In this scenario, the company needs a ticket management app - not a Todo app, so I make changes to the web source code and I don't need to provision any additional resources in the infrastructure.

I locate the header file. One way (and a slow one) of doing it is running a command in Powershell:
`Get-ChildItem -Recurse "header.tsx"`  



I then use Vim text editor to open the header.tsx file.



... and write "Ticket Manager" instead of "ToDo" in the header display text.



Then save and exit the Vim editor and run `azd deploy` command and go to the website to see the transformation on the webpage:
