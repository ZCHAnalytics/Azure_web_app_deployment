## 2. Monitoring Options

Azure Developer CLI templates also provisioned a monitoring resources `Application Insights`. This tool generates  dashboards with live metrics and logging. 

Dashboards can be launched with the `azd monitor` command and one of the following flags:

`--overview` - Launches the main dashboard.
`--live` - Launches the live metrics dashboard.
`--logs` - Launches the logging dashboard.
