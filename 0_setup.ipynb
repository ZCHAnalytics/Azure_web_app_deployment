# 1. Project Setup  

## 1.1. Project Prerequisites 
- Azure Developer command-line interface (CLI) for provisioning and deployment to Azure
- Node.js with npm (18.17.1+) for running a server-side API environment
- GitHub account to configure CI/CD pipelines
- Azure account for deloying and managing cloud resources  
- Choice of shell. I chose Window Powershell for command line tasks and Visual Studio Code's integrated Git Bash for visual feedback

## 1.2. Optional 
- Dedicated environment, e.gg. conda environment, as my termminal uses Node.js v20 and is not compatible.
- Text editor for changing code and configuration files. I chose VIM editor.  

## 2. Identify Project Template

Azure Developer CLI, `azd`, supports a number of templates with the necessary source code and configuration assets for provisioning and deploying applications to Azure. 

An existing app can be converted into an `azd` template with the following core steps:
- Initialize the project template with `azd init`
- Create the infrastructure as code files using Bicep or Terraform
- Update the `azure.yaml` file to tie the app services together with the Azure resources
- Provision & deploy with `azd up`



## 3. Initialise the Template 
I choose a template called `todo-nodejs-mongo`. It serves as a blueprint for getting a React web app with a Node.js API and a MongoDB database on Azure. I then run the following commands:

`azd init --template todo-nodejs-mongo`



`azd auth login` - to login into Azure portal (opens a web browser for sign in)
