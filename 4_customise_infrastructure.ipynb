## 4. Customising infrastructure code 

The ticket management app needs to provide customers with the option to upload screenshots and other files with their ticket. 

I need to provision a storage space to hold customer files. This requires the creation of an Azure Storage account. 

The IaaC folder `infra` contains various reusable Bicep modules, including one for storage. Lets check if it is there by running the following PowerShell command and copy the path to the file:
`Get-ChildItem -Recurse -Filter "*.bicep" | Where-Object { $_.Name -match 'storage'}`

It is indeed there. I take note of the file location to add to the main.bicep file.  

I then open the main.bicep file in a text editor, for instance, Vim, to add the storage module just before the outputs section in the end of the file.
